```python
# Создаём тестовый файл purchase_log.txt
test_data = """1840e0b9d4 Продукты
4e4f90fcfb Электроника
3a3b2c1d0e Одежда
5f6g7h8i9j Книги
2k3l4m5n6p Игрушки"""

with open('purchase_log.txt', 'w', encoding='utf-8') as file:
    file.write(test_data)

# Функция для создания словаря покупок
def create_purchases_dict(file_path):
    purchases = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            user_id, category = line.strip().split()
            purchases[user_id] = category
    return purchases

# Используем функцию
file_path = 'purchase_log.txt'
purchases = create_purchases_dict(file_path)

# Выводим первые два элемента словаря
print("Первые два элемента словаря purchases:")
count = 0
for user_id, category in purchases.items():
    if count < 2:
        print(f"{user_id} '{category}'")
        count += 1
    else:
        break

"""Сначала создаем тестовый файл с помощью команды test_data, затем эти данные записываются в файл purchase_log.txt. Функция create_purchases_dict читает файл и преобразует его в словарь"""

# Создаём словарь покупок из purchase_log.txt (пример данных)
purchases = {
    '1840e0b9d4': 'Продукты',
    '4e4f90fcfb': 'Электроника',
    '3a3b2c1d0e': 'Одежда',
    '5f6g7h8i9j': 'Книги',
    '2k3l4m5n6p': 'Игрушки'
}

# Имя входного файла с визитами
visit_log_file = 'visit_log.csv'
# Имя выходного файла с визитами, где были покупки
funnel_file = 'funnel.csv'

# Создаём тестовый файл visit_log.csv с примером данных
test_visit_data = """user_id,source
1840e0b9d4,other
4e4f90fcfb,context
7a8b9c0d1e,ads
3a3b2c1d0e,search
5f6g7h8i9j,email"""

# Записываем тестовые данные в файл visit_log.csv
with open(visit_log_file, 'w', encoding='utf-8') as file:
    file.write(test_visit_data)

# Открываем файл funnel.csv для записи результата
with open(funnel_file, 'w', encoding='utf-8') as funnel:
    # Записываем заголовок в файл
    funnel.write("user_id,source,category\n")

    # Открываем файл visit_log.csv для чтения
    with open(visit_log_file, 'r', encoding='utf-8') as visits:
        # Пропускаем заголовок (первую строку)
        next(visits)

        # Построчно читаем файл
        for line in visits:
            try:
                # Разбиваем строку на user_id и source
                user_id, source = line.strip().split(',')
                # Проверяем, есть ли user_id в словаре покупок
                if user_id in purchases:
                    # Записываем строку в funnel.csv с добавлением категории покупки
                    funnel.write(f"{user_id},{source},{purchases[user_id]}\n")
            except ValueError:
                # Если строка не разбивается на два значения, выводим ошибку
                print(f"Ошибка в строке: {line.strip()}")

print(f"Файл {funnel_file} успешно создан.")

"""Словарь purchases хранит соответствие user_id и категории покупки. Для проверки работы кода создаётся файл visit_log.csv с примером данных.
Чтение и запись файлов:

visit_log.csv читается построчно.
Для каждой строки проверяется наличие user_id в словаре purchases.
Если user_id найден, строка записывается в funnel.csv с добавлением категории покупки"""

```
